# TaiwanCredit Data

This example notebook demonstrates how to use PiML in its low-code mode for developing machine learning models for the TaiwanCredit data from UCI repository, which consists of 30,000 credit card clients in Taiwan from 200504 to 200509; see details [here](https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients). The data can be loaded from PiML and it is subject to slight preprocessing.

The response `FlagDefault` is binary and it is a classification problem.

Click the ipynb links to run examples in [Google Colab](https://colab.research.google.com/github/SelfExplainML/PiML-Toolbox/blob/main/examples/Example_TaiwanCredit.ipynb).

## Load and Prepare Data

```python
from piml import Experiment
exp = Experiment()
```

```python
# Choose TaiwanCredit
exp.data_loader()
```

```python
# Use only payment history attributes: Pay_1~6, BILL_AMT1~6 and PAY_AMT1~6
# Keep the response `FlagDefault`, while excluding all other variables
exp.data_summary()
```

```python
# Prepare dataset with default settings
exp.data_prepare()
```

```python
exp.eda()
```

## Train Intepretable Models



```python
# Choose EBM and ReLU-DNN; Customize ReLU-DNN with L1_regularization = 0.0008; then register each trained model.
exp.model_train()
```

## Interpretability and Explainability

```python
# Model-specific inherent interpretation including feature importance, main effects and pairwise interactions.
exp.model_interpret()
```

```python
# Model-agnostic post-hoc explanation by Permutation Feature Importance, PDP (1D and 2D) vs. ALE (1D and 2D), LIME vs. SHAP
exp.model_explain()
```

## Model Diagnostics and Outcome Testing

```python
exp.model_diagnose()
```

## Model Comparison and Benchmarking

```python
exp.model_compare()
```
